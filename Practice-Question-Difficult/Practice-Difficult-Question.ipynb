```python
# 1. In an organisation monthly sales figures of employees are available. Write a program to find the total sales of each employee and total sales of each month. Also find the total sales of the organization.

import numpy as np

# Problem 1: Monthly Sales Analysis
def sales_analysis(sales_data):
    sales_data = np.array(sales_data)
    total_sales_per_employee = np.sum(sales_data, axis=1)
    total_sales_per_month = np.sum(sales_data, axis=0)
    total_sales_organization = np.sum(sales_data)
    
    return total_sales_per_employee, total_sales_per_month, total_sales_organization

# Example Usage
sales_data = [
    [1000, 1500, 2000],  # Employee 1
    [1200, 1300, 1700],  # Employee 2
    [900, 1100, 1600]    # Employee 3
]
total_sales_per_employee, total_sales_per_month, total_sales_organization = sales_analysis(sales_data)

print("Total Sales per Employee:", total_sales_per_employee)
print("Total Sales per Month:", total_sales_per_month)
print("Total Sales of Organization:", total_sales_organization)

# 2. Write a program to find the determinant of a matrix.

def matrix_determinant(matrix):
    return np.linalg.det(np.array(matrix))

# Example Usage
sales_data = [
    [1000, 1500, 2000],  # Employee 1
    [1200, 1300, 1700],  # Employee 2
    [900, 1100, 1600]    # Employee 3
]
total_sales_per_employee, total_sales_per_month, total_sales_organization = sales_analysis(sales_data)

print("Total Sales per Employee:", total_sales_per_employee)
print("Total Sales per Month:", total_sales_per_month)
print("Total Sales of Organization:", total_sales_organization)

# Example Usage for Determinant Calculation
matrix = [[2, 3], [1, 4]]
det_value = matrix_determinant(matrix)
print("Determinant of the matrix:", det_value)

# 3. Write a program to find the sum of diagonal elements of a square matrix.

def sum_diagonal(matrix):
    return np.trace(np.array(matrix))

# 4. Maximum temperature of each day for 20 cities are recorded for the month of January. Write a program to find the following:
#    a. The highest temperature of each city in January
#    b. The day in which the highest temperature is recorded for the city
#    c. Average maximum temperature of each city.

def temperature_analysis(temp_data):
    temp_data = np.array(temp_data)
    highest_temp_per_city = np.max(temp_data, axis=1)
    day_of_highest_temp = np.argmax(temp_data, axis=1) + 1  # Days are 1-based index
    avg_temp_per_city = np.mean(temp_data, axis=1)
    
    return highest_temp_per_city, day_of_highest_temp, avg_temp_per_city

# Example Usage
sales_data = [
    [1000, 1500, 2000],  # Employee 1
    [1200, 1300, 1700],  # Employee 2
    [900, 1100, 1600]    # Employee 3
]
total_sales_per_employee, total_sales_per_month, total_sales_organization = sales_analysis(sales_data)

print("Total Sales per Employee:", total_sales_per_employee)
print("Total Sales per Month:", total_sales_per_month)
print("Total Sales of Organization:", total_sales_organization)

# Example Usage for Determinant Calculation
matrix = [[2, 3], [1, 4]]
det_value = matrix_determinant(matrix)
print("Determinant of the matrix:", det_value)

# Example Usage for Sum of Diagonal Elements
square_matrix = [[4, 2, 1], [3, 5, 7], [8, 6, 9]]
diag_sum = sum_diagonal(square_matrix)
print("Sum of diagonal elements:", diag_sum)

# Example Usage for Temperature Analysis
temp_data = np.random.randint(20, 40, size=(20, 31))  # Random temperatures for 20 cities, 31 days
highest_temp, highest_temp_day, avg_temp = temperature_analysis(temp_data)
print("Highest temperature per city:", highest_temp)
print("Day of highest temperature per city:", highest_temp_day)
print("Average maximum temperature per city:", avg_temp)


# 5. 5. Given an m x n matrix, if an element in the matrix is 0, set its entire row and column to 0. Do this in O(m * n) time complexity.
# Output:
# [ [1, 0, 1],
#  [0, 0, 0],
# [1, 0, 1] ]

# Input:
#  [ [1, 1, 1],
#   [1, 0, 1],
#   [1, 1, 1] ]
def set_zeroes(matrix):
    rows, cols = len(matrix), len(matrix[0])
    row_zero = set()
    col_zero = set()
    
    for i in range(rows):
        for j in range(cols):
            if matrix[i][j] == 0:
                row_zero.add(i)
                col_zero.add(j)
    
    for i in row_zero:
        matrix[i] = [0] * cols
    
    for j in col_zero:
        for i in range(rows):
            matrix[i][j] = 0
    
    return matrix

# Example Usage
sales_data = [
    [1000, 1500, 2000],  # Employee 1
    [1200, 1300, 1700],  # Employee 2
    [900, 1100, 1600]    # Employee 3
]
total_sales_per_employee, total_sales_per_month, total_sales_organization = sales_analysis(sales_data)

print("Total Sales per Employee:", total_sales_per_employee)
print("Total Sales per Month:", total_sales_per_month)
print("Total Sales of Organization:", total_sales_organization)

# Example Usage for Determinant Calculation
matrix = [[2, 3], [1, 4]]
det_value = matrix_determinant(matrix)
print("Determinant of the matrix:", det_value)

# Example Usage for Sum of Diagonal Elements
square_matrix = [[4, 2, 1], [3, 5, 7], [8, 6, 9]]
diag_sum = sum_diagonal(square_matrix)
print("Sum of diagonal elements:", diag_sum)

# Example Usage for Temperature Analysis
temp_data = np.random.randint(20, 40, size=(20, 31))  # Random temperatures for 20 cities, 31 days
highest_temp, highest_temp_day, avg_temp = temperature_analysis(temp_data)
print("Highest temperature per city:", highest_temp)
print("Day of highest temperature per city:", highest_temp_day)
print("Average maximum temperature per city:", avg_temp)

# Example Usage for Setting Matrix Zeroes
input_matrix = [
    [1, 1, 1],
    [1, 0, 1],
    [1, 1, 1]
]
output_matrix = set_zeroes(input_matrix)
print("Modified Matrix:", output_matrix)
```
